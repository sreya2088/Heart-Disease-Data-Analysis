```python
df_cleaned = df.dropna(subset=['TotalWorkingYears'])

plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['TotalWorkingYears'], bins=20, kde=True)
plt.title('Distribution of Total Working Years')
plt.xlabel('Total Working Years')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Attrition'] = df['Attrition'].map({'No': 0, 'Yes': 1})
```

```python
attrition_rate = df.groupby('Department')['Attrition'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Department', y='Attrition', data=attrition_rate)
plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
attrition_rate_job_role = df.groupby('JobRole')['Attrition'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='JobRole', y='Attrition', data=attrition_rate_job_role)
plt.title('Attrition Rate by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Attrition Rate')
plt.xticks(rotation=45)
plt.show()
```
